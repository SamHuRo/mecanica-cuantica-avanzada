Compruebe que $|r_{ba}|^2$ cumple con las reglas de selección $\Delta \ell = \pm 1$, $\Delta m = \pm 1, 0$

Calcule $\braket{\psi_{a}|x |\psi_{b}}$, $\braket{\psi_{a}|y | \psi_{b}}$, $ \braket{\psi_{a}|z |\psi_{b}}$ y eleve al cuadrado. 

Tambien se puede ver a partir de la expresión $(1)$: 
$$
\int Y_{l_1 m_1}(\theta,\varphi)\,
     Y_{l_2 m_2}(\theta,\varphi)\,
     Y_{l_3 m_3}(\theta,\varphi)\,
     \sin\theta \, d\theta\, d\varphi
=
\sqrt{
\frac{(2l_1+1)(2l_2+1)(2l_3+1)}{4\pi}
}
\begin{pmatrix}
l_1 & l_2 & l_3 \\
0   & 0   & 0
\end{pmatrix}
\begin{pmatrix}
l_1 & l_2 & l_3 \\
m_1 & m_2 & m_3
\end{pmatrix}
$$  



Con: 

* $x \approx Y_{1,1} - Y_{1,-1} $

* $y \approx Y_{1,1} + Y_{1,-1} $

* $z \approx Y_{1,0} $


```python
import sympy as sp
from sympy import simplify
from sympy.physics.hydrogen import Psi_nlm, R_nl
```

```python
#  Definición de las constantes
factor_conversion = {
    "cm": 219474.63137,
    "eV": 27.2114
}

# Definición de las constantes
alpha = 1 / 137.035999
c = 1 / alpha
Z = 1               # Número atomico
mu = 1              # Masa reducida
me = 1             # Masa del electron

# Definición de los numeros cuanticos
n_max = 2
```

```python
# Definición de la variables simbólicas
r, theta, phi = sp.symbols('r theta phi', real=True)
hbar = sp.symbols('hbar', real=True, positive=True)
```

```python
# Coordenadas cartecianas en coordenadas esfericas
x = r * sp.sin(theta) * sp.cos(phi)
y = r * sp.sin(theta) * sp.sin(phi)
z = r * sp.cos(theta)

# Calculo de los elementos de la matriz
for n in range(1, n_max + 1):
    for n_p in range(1, n_max + 1):
        for l in range(0, n):  # noqa: E741
            for lp in range(0, n_p):
                for ml in range(-l, l + 1):
                    for mlp in range(-lp, lp + 1):
                        psi_conjugada = sp.conjugate(Psi_nlm(n, l, ml, r, phi, theta))
                        psi = Psi_nlm(n_p, lp, mlp, r, phi, theta)
                        xab = sp.integrate(
                            psi_conjugada * x * psi * r**2 * sp.sin(theta), 
                            (r, 0, sp.oo), 
                            (theta, 0, sp.pi), 
                            (phi, 0, 2 * sp.pi)
                        )
                        yab = sp.integrate(
                            psi_conjugada * y * psi * r**2 * sp.sin(theta), 
                            (r, 0, sp.oo), 
                            (theta, 0, sp.pi), 
                            (phi, 0, 2 * sp.pi)
                        )
                        zab = sp.integrate(
                            psi_conjugada * z * psi * r**2 * sp.sin(theta), 
                            (r, 0, sp.oo), 
                            (theta, 0, sp.pi), 
                            (phi, 0, 2 * sp.pi)
                        )
                        rab = sp.Abs(xab) ** 2 + sp.Abs(yab) ** 2 + sp.Abs(zab) ** 2
                        print(f"<psi_({n},{l},{ml}) | psi_({n_p},{lp},{mlp})> = {simplify(rab).n():.4f}, delta_l = {lp - l}, delta_ml = {mlp - ml}")
                            
```

```python
# También se puede utilizar los coefiecientes de la ecuación (1)
from sympy.physics.wigner import gaunt

for n in range(1, n_max + 1):
    for n_p in range(1, n_max + 1):
        for l in range(0, n):  # noqa: E741
            for lp in range(0, n_p):
                for ml in range(-l, l + 1):
                    for mlp in range(-lp, lp + 1):
                        integral_r = sp.integrate(
                            r**3 * sp.conjugate(R_nl(n, l, r)) * R_nl(n_p, lp, r), 
                            (r, 0, sp.oo)
                        )
                        xab = (
                            (gaunt(l, 1, lp, ml, 1, mlp) - gaunt(l, 1, lp, ml, -1, mlp)) * 
                            integral_r * 
                            sp.sqrt((2 * sp.pi) / 3) * (-1) ** ml
                        )
                        yab = (
                            (gaunt(l, 1, lp, ml, 1, mlp) + gaunt(l, 1, lp, ml, -1, mlp)) * 
                            integral_r * 
                            sp.sqrt((2 * sp.pi) / 3) * (-1) ** (ml) * 1j
                        )
                        zab = (
                            (gaunt(l, 1, lp, ml, 0, mlp)) * 
                            integral_r * 
                            sp.sqrt(4 * sp.pi / 3) * (-1) ** ml
                        )
                        rab = sp.Abs(xab.n()) ** 2 + sp.Abs(yab.n()) ** 2 + sp.Abs(zab.n()) ** 2
                        print(f"<psi_({n},{l},{ml}) | psi_({n_p},{lp},{mlp})> = {rab:.4f}, delta_l = {lp - l}, delta_ml = {mlp - ml}")

```
